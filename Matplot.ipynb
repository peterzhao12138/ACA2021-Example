```python
import csv
all_result = []
with open('2048.csv', newline='') as csvfile:
  spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
  for row in spamreader:
    all_result.append(row[0].split(','))
print(all_result[0])
```

```python
data = [[row[5] for row in all_result][1:],
[row[6] for row in all_result][1:],
[row[7] for row in all_result][1:],
[row[8] for row in all_result][1:],
[row[9] for row in all_result][1:],
[row[10] for row in all_result][1:]]
```
